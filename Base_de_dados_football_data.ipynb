```python
lista_dfs =[]
```

```python
for url in url_csv:
    df = pd.read_csv(url)
    lista_dfs.append(df)
```

```python
df = pd.concat(lista_dfs, ignore_index=True)
```

```python
df = df[['Date' , 'HomeTeam' , 'AwayTeam' , 'FTHG' , 'FTAG' , 'FTR' , 'B365H' , 'B365D' , 'B365A']]
```

```python
df.columns = ['Data' , 'Home' , 'Away' , 'HG' , 'AG' , 'Result' , 'H_odd' , 'D_odd' , 'A_odd']
```

```python
#Criando coluna 'Pts_H

df.loc[(df['HG']  > df['AG']), 'Pts_H'] = 3
df.loc[(df['HG'] == df['AG']), 'Pts_H'] = 1
df.loc[(df['HG']  < df['AG']), 'Pts_H'] = 0

#Criando coluna 'Pts_A'
df.loc[(df['HG']  < df['AG']), 'Pts_A'] = 3
df.loc[(df['HG'] == df['AG']), 'Pts_A'] = 1
df.loc[(df['HG']  > df['AG']), 'Pts_A'] = 0
```

```python
#Criando função para limpar o df

def drop_reset_index(df):
    df = df.dropna().reset_index(drop=True)
    df.index+1
    return df
```

```python
#Criando as variáveis
```

```python
#Probabilidade

df['pro_H'] = 1/ df['H_odd']
df['pro_D'] = 1/ df['A_odd']
df['pro_A'] = 1/ df['A_odd']

#CV das Odds
desv_pad = df[['H_odd' , 'D_odd' , 'A_odd']].std(axis=1)
med_odd = df[['H_odd' , 'D_odd' , 'A_odd']].mean(axis=1)
cv_odd = desv_pad / med_odd
df['CV_odd'] = cv_odd

 
#CV das Probabilidades
desv_pad = df[['pro_H' , 'pro_D' , 'pro_A']].std(axis=1)
med_pro = df[['pro_H' , 'pro_D' , 'pro_A']].mean(axis=1)
cv_pro = desv_pad / med_pro
df['CV_pro'] = cv_pro
```

```python
# Média de pontos

df['Med_pts_H'] = df.groupby('Home')['Pts_H'].rolling(window=5,min_periods=5).mean().reset_index(0 , drop=True)
df['Med_pts_H'] = df.groupby('Home')['Med_pts_H'].shift(1)

df['Med_pts_A'] = df.groupby('Away')['Pts_A'].rolling(window=5,min_periods=5).mean().reset_index(0 , drop=True)
df['Med_pts_A'] = df.groupby('Away')['Med_pts_A'].shift(1)

# Desvio padrão dos pontos

df['Desvpad_pts_H'] = df.groupby('Home')['Pts_H'].rolling(window=5,min_periods=5).std().reset_index(0 , drop=True)
df['Desvpad_pts_H'] = df.groupby('Home')['Desvpad_pts_H'].shift(1)

df['Desvpad_pts_A'] = df.groupby('Away')['Pts_A'].rolling(window=5,min_periods=5).std().reset_index(0 , drop=True)
df['Desvpad_pts_A'] = df.groupby('Away')['Desvpad_pts_A'].shift(1)

# CV da média dos pontos

df['CV_pts_H'] = df['Desvpad_pts_H'] / df['Med_pts_H']
df['CV_pts_A'] = df['Desvpad_pts_A'] / df['Med_pts_A']

# Média de gols marcados 

df['Média_GM_H'] = df.groupby('Home')['HG'].rolling(window=5,min_periods=5).mean().reset_index(0 , drop=True)
df['Média_GM_H'] = df.groupby('Home')['Média_GM_H'].shift(1)

df['Média_GM_A'] = df.groupby('Away')['AG'].rolling(window=5,min_periods=5).mean().reset_index(0 , drop=True)
df['Média_GM_A'] = df.groupby('Away')['Média_GM_A'].shift(1)

# Desvio padrão dos gols marcados

df['Desvpad_GM_H'] = df.groupby('Home')['HG'].rolling(window=5,min_periods=5).std().reset_index(0 , drop=True)
df['Desvpad_GM_H'] = df.groupby('Home')['Desvpad_GM_H'].shift(1)

df['Desvpad_GM_A'] = df.groupby('Away')['AG'].rolling(window=5,min_periods=5).std().reset_index(0 , drop=True)
df['Desvpad_GM_A'] = df.groupby('Away')['Desvpad_GM_A'].shift(1)

# CV Gols marcados

df['CV_GM_H'] = df['Desvpad_GM_H'] / df['Média_GM_H']
df['CV_GM_A'] = df['Desvpad_GM_A'] / df['Média_GM_A']

# Média de gols sofridos

df['Média_GS_H'] = df.groupby('Home')['AG'].rolling(window=5,min_periods=5).mean().reset_index(0 , drop=True)
df['Média_GS_H'] = df.groupby('Home')['Média_GS_H'].shift(1)

df['Média_GS_A'] = df.groupby('Away')['HG'].rolling(window=5,min_periods=5).mean().reset_index(0 , drop=True)
df['Média_GS_A'] = df.groupby('Away')['Média_GS_A'].shift(1)

# Desvio padrão de gols sofridos

df['Desvpad_GS_H'] = df.groupby('Home')['AG'].rolling(window=5,min_periods=5).std().reset_index(0 , drop=True)
df['Desvpad_GS_H'] = df.groupby('Home')['Desvpad_GS_H'].shift(1)

df['Desvpad_GS_A'] = df.groupby('Away')['HG'].rolling(window=5,min_periods=5).std().reset_index(0 , drop=True)
df['Desvpad_GS_A'] = df.groupby('Away')['Desvpad_GS_A'].shift(1)

# CV dos gols sofridos

df['CV_GS_H'] = df['Desvpad_GS_H'] / df['Média_GS_H']
df['CV_GS_A'] = df['Desvpad_GS_A'] / df['Média_GS_A']

# Saldo de gols

df['SG_H'] = df['HG'] - df['AG']
df['SG_A'] = df['AG'] - df['HG']

# Média do Saldo de gols

df['Média_SG_H'] = df.groupby('Home')['SG_H'].rolling(window=5,min_periods=5).mean().reset_index(0 , drop=True)
df['Média_SG_H'] = df.groupby('Home')['Média_SG_H'].shift(1)

df['Média_SG_A'] = df.groupby('Away')['SG_A'].rolling(window=5,min_periods=5).mean().reset_index(0 , drop=True)
df['Média_SG_A'] = df.groupby('Away')['Média_SG_A'].shift(1)

# Desvio padrão do Saldo de gols

df['Desvpad_SG_H'] = df.groupby('Home')['SG_H'].rolling(window=5,min_periods=5).std().reset_index(0 , drop=True)
df['Desvpad_SG_H'] = df.groupby('Home')['Desvpad_SG_H'].shift(1)

df['Desvpad_SG_A'] = df.groupby('Away')['SG_A'].rolling(window=5,min_periods=5).std().reset_index(0 , drop=True)
df['Desvpad_SG_A'] = df.groupby('Away')['Desvpad_SG_A'].shift(1)

# CV do Saldo de gols

df['CV_SG_H'] = df['Desvpad_SG_H'] / df['Média_SG_H']
df['CV_SG_A'] = df['Desvpad_SG_A'] / df['Média_SG_A']

# Custo do gol

df['CG_H'] = df['HG'] / df['pro_H']
df['CG_A'] = df['AG'] / df['pro_A']

# Média do custo do gol

df['Média_CG_H'] = df.groupby('Home')['CG_H'].rolling(window=5,min_periods=5).mean().reset_index(0 , drop=True)
df['Média_CG_H'] = df.groupby('Home')['Média_CG_H'].shift(1)

df['Média_CG_A'] = df.groupby('Away')['CG_A'].rolling(window=5,min_periods=5).mean().reset_index(0 , drop=True)
df['Média_CG_A'] = df.groupby('Away')['Média_CG_A'].shift(1)

# Desvio padrão do custo do gol

df['Desvpad_CG_H'] = df.groupby('Home')['CG_H'].rolling(window=5,min_periods=5).std().reset_index(0 , drop=True)
df['Desvpad_CG_H'] = df.groupby('Home')['Desvpad_CG_H'].shift(1)

df['Desvpad_CG_A'] = df.groupby('Away')['CG_A'].rolling(window=5,min_periods=5).std().reset_index(0 , drop=True)
df['Desvpad_CG_A'] = df.groupby('Away')['Desvpad_CG_A'].shift(1)

# CV do custo do gol

df['CV_CG_H'] = df['Desvpad_CG_H'] / df['Média_CG_H']
df['CV_CG_A'] = df['Desvpad_CG_A'] / df['Média_CG_A']

#Limpeza das colunas
colunas_deletar = [ 'Pts_H', 'Pts_A', 'Desvpad_pts_H', 'Desvpad_pts_A', 'Desvpad_GM_H', 'Desvpad_GM_A','Desvpad_GS_H', 'Desvpad_GS_A',
                   'Desvpad_SG_H', 'Desvpad_SG_A', 'Desvpad_CG_H', 'Desvpad_CG_A',  'CG_H', 'CG_A' ,  'SG_H', 'SG_A']
df = df.drop(colunas_deletar , axis=1)
df = drop_reset_index(df)
```

```python
# Removendo inf

df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Removendo NaN

df = df.dropna()
```

```python
# Arredondando para duas casas depois da vírgula
df = df.round(2)
```
